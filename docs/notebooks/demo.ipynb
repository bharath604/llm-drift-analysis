```python
import pandas as pd

# Load sample dataset
df = pd.read_csv("docs/data/sample.csv")
df.head()
```

```python
import matplotlib.pyplot as plt

# Compare ground truth vs predictions
counts = df.groupby(["ground_truth","prediction"]).size().unstack(fill_value=0)

counts.plot(kind="bar", figsize=(6,4))
plt.title("Ground Truth vs Predictions")
plt.ylabel("Count")
plt.show()
```
